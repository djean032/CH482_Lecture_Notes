# Finite Potentials and Quantum Tunneling

## Finite Potentials

### Step Potential
First, let's plot what we mean by finite potential.

```python
import numpy as np 
import matplotlib.pyplot as plt

# Create a plot of partile in a box potentials with finite potential
x = np.linspace(-1, 1, 1000)
V = np.zeros_like(x)
V[500:1000] = 0.5

# label y axis as V(x) with potential as V_0
plt.ylabel(r'$V(x)$')
plt.xlabel(r'$x$', loc='right')
plt.ylim(-0.1, .7)
plt.title(r'Free Particle with Step Potential')
# label the regions
plt.text(-0.75, 0.3, "Region 1\n" r"$V(x) = 0$")
plt.text(0.35, 0.3, "Region 2\n" r"$V(x) = V_0$")
# No numbers on either axis
plt.xticks([0], ['0'])
plt.yticks([])
plt.vlines(x[500], -0.1, 1, linestyle='--', color='black')
plt.plot(x, V)

#
```

From the above we can see we mean a potential that is some set height, here depcited by $V_0$.


Now let's talk about our wavefunctions for these two regions.

Region 1:

$$\psi_1(x) = Ae^{ik_1x} + Be^{-ik_1x}$$


Region 2:

$$\psi_1(x) = Ce^{ik_2x} + De^{-ik_2x}$$

where $e^{ik_1x}$ is the particle moving right and $e^{-ik_1x}$ is a particle moving left and

$$
k_1 = \left(\frac{2mE}{\hbar^2}\right)^{1/2} \\~\\
k_2 = \left(\frac{2m(E-V_0)}{\hbar^2}\right)^{1/2}
$$

We can determine the probability that a particle is travelling in a certain direction using the coefficients. For instance, ${\left\lvert A\right\rvert}^2$ is the probability that the particle is traveling with a momentum of $\hbar k_1$ in Region 1. 

Additionally, we can extend this to many particles, $N_0$, giving ${\lvert A \rvert}^2 N_0$, for the number of particles traveling with momentum $\hbar k_1$. We can also classify how many particles pass a point in a given time by using the velocity, given by $v=\frac{\hbar k_1}{m}$. This gives us $v{\left\lvert A \right\rvert}^2 N_0$.

Let's ignore the case where a particle is in Region 2, moving left, so $D=0$. Now, applying our conditions that $\psi(x)$ and $\frac{d\psi}{dx}$ must be continuous at $x=0$, we obtain:

$$
A+B = C \\~\\
k_1(A-B) = k_2C
$$

Now we can define a quantity, $R$, which is the probability that a particle is reflected by our barrier as given by:

$$
\begin{align}
R &= \frac{v_1 {\lvert B \rvert}^2N_0}{v_1 {\lvert A \rvert}^2 N_0} \\
&= \frac{\hbar k_1 {\lvert B \rvert}^2N_0/m}{\hbar k_1 {\lvert A \rvert}^2 N_0/m} \\
&= \frac{{\lvert B \rvert}^2}{{\lvert A \rvert}^2}
\end{align}
$$

Now, doing the requisite algebra, we can show that:

$$
R = \left(\frac{k_1 - k_2}{k_1+k_2}\right)^2
$$
Similarly,

$$
\begin{align}
T &= \frac{v_2 {\lvert C \rvert}^2N_0}{v_1 {\lvert A \rvert}^2 N_0} \\
&= \frac{\hbar k_2 {\lvert C \rvert}^2N_0/m}{\hbar k_1 {\lvert A \rvert}^2 N_0/m} \\
&= \frac{k_2{\lvert C \rvert}^2}{k_1{\lvert A \rvert}^2}
\end{align}
$$

From which, again with nasty algebra, we see:

$$
T =\frac{4k_1k_2}{\left(k_1+k_2\right)^2}
$$



which defines the probability that a particle "tunnels" through the barrier. These two coefficients are called the Reflection and Transmission coefficients respectively.

### Rectangular Potential
Again, let's do a plot.












```python
import numpy as np 
import matplotlib.pyplot as plt

# Create a plot of partile in a box potentials with finite potential
x = np.linspace(-1, 1, 1000)
V = np.zeros_like(x)
V[500:700] = 0.5

# label y axis as V(x) with potential as V_0
plt.ylabel(r'$V(x)$')
plt.xlabel(r'$x$', loc='right')
plt.ylim(-0.1, .7)
plt.title(r'Free Particle with Rectangular Potential')
# label the regions
plt.text(-0.75, 0.3, "Region 1\n" r"$V(x) = 0$")
plt.text(.075, 0.3, "Region 2\n" r"$V(x) = V_0$")
plt.text(0.6, 0.3, "Region 3\n" r"$V(x) = 0$")
# No numbers on either axis
plt.xticks([0, 0.4], ['0', 'a'])
plt.yticks([])
plt.vlines(x[500], -0.1, 1, linestyle='--', color='black')
plt.vlines(x[700], -0.1, 1, linestyle='--', color='black')
plt.plot(x, V)

#
```

Now let's talk about our wavefunctions for these three regions.

Region 1 first.

$$\psi_1(x) = Ae^{ik_1x} + Be^{-ik_1x}$$


and similarly for Regions 2 and 3:

$$\psi_2(x) = Ce^{k_2x} + De^{-k_2x}\\~\\
\psi_3(x) = Ee^{ik_1x} + Fe^{-ik_1x}$$

where $e^{ik_1x}$ is the particle moving right and $e^{-ik_1x}$ is a particle moving left and

$$
k_1 = \left(\frac{2mE}{\hbar^2}\right)^{1/2} \\~\\
k_2 = \left(\frac{2m(E-V_0)}{\hbar^2}\right)^{1/2}
$$

Similarly to the step potential, we can apply our continutity requirements of $\psi(x)$ and $\frac{d\psi}{dx}$ at $x=0$ and $x=a$ and we get four more equations.
$$
A+B=C+D \\~\\
ik_1(A-B) = k_2(C-D)\\~\\
Ce^{k^2a} + De^{-k_2a} = Ee^{ik_1a} \\~\\
k_2Ce^{k_2a} - k_2De^{-k_2a} = ik_1Ee^{ik_1a}
$$

The algebra for this is quite tedious, so let's use Sympy to do this one.

First, let's solve our equations. Remember from your algebra classes, we have 4 equations and 5 unknowns, so we will put one of the coefficients that we don't need for our two relationships last, so each of the other coefficients will be solved for in terms of that coefficient. In this case, we will not determine $D$.


```python
import sympy as sy
A, B, C, D, E = sy.symbols('A B C D E')
k1, k2, a = sy.symbols('k1 k2 a', real=True)
hbar, E, V_0, m = sy.symbols('hbar E V_0 m', real=True)
eq1 = sy.Eq(A + B, C + D)
eq2 = sy.Eq(sy.I*k1*(A - B), k2*(C - D))
eq3 = sy.Eq(C*sy.exp(k2*a) + D*sy.exp(-k2*a), E*sy.exp(sy.I*k1*a))
eq4 = sy.Eq(k2*C*sy.exp(k2*a) - k2*D*sy.exp(-k2*a), sy.I * k1* E*sy.exp(sy.I*k1*a))
solution = sy.solve([eq1, eq2, eq3, eq4], (A, E, B, C, D))
print(solution)
```

Now we have A, B, and E solved for in terms of D, so we can solve for our relationships. Let's solve for T first and evaluate it numerically with some fake values to compare to the analytic expression from the text.

```python
expr1 = solution[A]
expr2 = solution[E]
expr3 = (expr2 / expr1).simplify()
expr3_conj = sy.conjugate(expr3)
expr3_comb = expr3*expr3_conj
expr4 = (expr3_comb).simplify()
T = expr4.subs({k1: sy.sqrt(2*m*E/hbar**2), k2: sy.sqrt(2*m*(V_0-E)/hbar**2)})
T.subs({E: 0.1, V_0: 0.5, m: 1, hbar: 1, a: 1}).evalf()
```

Alright, let's determine if this is correct by evaluating the expression from the text with the same values.

```python
expr6 = 4/(4+((k1**2 + k2**2)**2/(k1**2*k2**2)) * sy.sinh(k2*a)**2)
expr7 = expr6.subs({k1: sy.sqrt(2*m*E/hbar**2), k2: sy.sqrt(2*m*(V_0 - E)/hbar**2)}).simplify()
expr7.subs({E: 0.1, V_0: 0.5, m: 1, hbar: 1, a: 1}).evalf()
```

Excellent! We got the right answer.

Now, what do we think happens if $E>V_0$? Would it always tunnel? Let's check.

```python
t_eval=T.subs({E: 1, V_0: 0.5, m: 1, hbar: 1, a: 1}).evalf()
t_eval
```

It's not 1! That means we have a chance that it's reflected back.

Let's now do the same computation for R.

```python
expr1 = solution[A]
expr8 = solution[B]
expr9 = (expr8 / expr1).simplify()
expr9_conj = sy.conjugate(expr9)
expr9_comb = expr9*expr9_conj
expr10 = (expr9_comb).simplify()
R = expr10.subs({k1: sy.sqrt(2*m*E/hbar**2), k2: sy.sqrt(2*m*(V_0-E)/hbar**2)})
r_eval = sy.Abs(R.subs({E: 1, V_0: 0.5, m: 1, hbar: 1, a: 1}).evalf())
r_eval
```

Now, let's add the Transmission probability and the Reflection probability. What do you expect?

```python
t_eval+r_eval
```

Excellent. The particle either tunnels or it is reflected, so the probability of both is 1. 

$$T+R=1$$

Let's plot $T$ vs $E/V_0$ and $a$ to see how tunneling is affected by the energy barrier and the size of the "box".

```python
import matplotlib.pyplot as plt
import numpy as np
def expr_T(E, V_0, m, hbar, a):
    return 1/(1+np.sinh(np.sqrt(2*m*a**2*V_0/hbar**2)*np.sqrt(1-E/V_0))**2/((4*E/V_0)*(1-E/V_0)))
def expr_T_sin(E, V_0, m, hbar, a):
    return 1/(1+np.sin(np.sqrt(2*m*a**2*V_0/hbar**2)*np.sqrt(E/V_0-1))**2/((4*E/V_0)*(E/V_0 - 1)))
V_0 = 1
a = 2.2
hbar = 1
m = 1
E1 = np.linspace(0.0001, 1, 200)
E2 = np.linspace(1, 4.9999, 800)
T_evald = expr_T(E1, V_0, m, hbar, a)
T2_evald = expr_T_sin(E2, V_0, m, hbar, a)
fig, ax = plt.subplots()
ax.plot(E1, T_evald)
ax.plot(E2, T2_evald)
ax.set_xlabel(r'$E$')
ax.set_ylabel(r'$T$')
```
