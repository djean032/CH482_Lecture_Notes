# Postulates of Quantum Mechanics

## The State of a System is Completely Specified by its Wave Function

### Postulate 1
In other words, all the dynamical information of a quantum system is contained within the wave function, with those dynamical variables being called the observables.

Due to this definition, and given the definition that

$$ \psi^*(x,y,z)\psi(x,y,z)dxdydz$$

is the probability that a particle lies in the volume element $dxdydz$. To satisfy the physical requirement, that a particle exists, we require that

$$ \int_{-\infty}^{\infty}\psi^*(x,y,z)\psi(x,y,z)dxdydz = 1$$

This can also be stated that the probability that we will find a particle if we look over all space is $100\%$.

Additionally, to be physically suitable wave function, it must be well behaved. Its integral must converge (over the interval in question) and its first derivative must be single-valued, continuous, and finite.

## Observables are Eigenvalues of Quantum Mechanical Operators

### Postulate 3
During the measurement of an observable, the only values that will actually be observed are those values, $a_n$, that satisfy the following expression:

$$\hat{A}\psi_n=a_n\psi_n$$


### Postulate 4
A system in a state described by a normalized wavefunction, $\psi$, has expectation values for its observables defined as:

$$\langle a \rangle = \int_{-\infty}^{\infty}\psi^*\hat{A}\psi dx$$


Let's give a quick example of postulates 3 and 4. Let's suppose that:

$$ \hat{A}\psi _n(x)=a_n\psi _n(x)$$

Thus we get

$$\langle a \rangle = \int_{-\infty}^{\infty}\psi^*\hat{A}\psi dx = \int_{-\infty}^{\infty}\psi^*a_n\psi dx=\int_{-\infty}^{\infty}\psi^*\psi dx= a_n$$

Let's also evalue the following

$$\hat{A}^2\psi_n(x)=\hat{A}[\hat{A}\psi_n(x)]=\hat{A}[a_n\psi_n(x)]=a_n[\hat{A}\psi_n(x)] = a_n^2\psi_n(x)$$

Now, we can calculate the variance of the measurements of $a$.

$$\sigma_a^2= \langle a^2 \rangle - {\langle a \rangle}^2 = a_n^2 - a_n^2 = 0$$

This demonstrates that the only value for the observable $a$ is the value $a_n$ that corresponds to the eigenfunction $\psi_n$. Additionally, we see that any observable must be **REAL**. Duh, it's a physical measurement! This does **NOT** mean the operator must completely real, it can have complex components.

### Postulate 5
The wavefunction evolves in time according to the time-dependent Schrodinger equation.

$$ \hat{H}\Psi(x,t)=i\hbar\frac{\partial\Psi(x,t)}{\partial t}$$

This is the first postulate that is new. $\hat{H}$ does not contain time, so we are able to separate their variables.

$$\Psi(x, t)=\psi(x)f(t)$$

Let's substitute this expression into the Time-Dependent Schrodinger Equation and divide through by $\psi(x)f(t)$:

$$\frac{1}{\psi(x)}\hat{H}\psi(x)=\frac{i\hbar}{f(t)}\frac{df(t)}{dt}$$

The left side is a function of x only and the right is only a function of t; therefore, they must each equal a constant. We'll denote this constant that relates each side as $E$, giving us the time-independent Schrodinger equation.

$$\hat{H}\psi(x)=E\psi(x)$$

and

$$\frac{df(t)}{dt}=-\frac{i}{\hbar}Ef(t)$$

Next, let us integrate the previous equation giving:

$$f(t)=e^{-iEt/\hbar}$$

Therefore, we can get the form of $\Psi(x, t)$ as

$$\Psi(x, t) = \psi(x)e^{-iEt/\hbar}$$

sometimes written as 

$$\Psi(x, t) = \psi(x)e^{-i\omega t}$$

Much of the time in chemistry we do not care about the time-dependent portion, assuming the system has an eigenstate, $\psi_n(x)$

$$\Psi_n^*(x, t)\Psi_n(x, t)dx = \psi_n^*(x)e^{iE_nt/\hbar}\psi_n(x)e^{-iE_nt/\hbar}dx=\psi_n^*(x)\psi_n(x)dx$$

Where our time-dependence disappears. These eigenstates are called *stationary states*.


### Orthonormality

We've explored the idea of the normalization of wavefunction previously. Now we contend with the fact that our eigenvalues are real. This means they must satisfy the following condition.

$$\int^{\infty}_{-\infty}\psi_m^*(x)\psi_n(x)dx=0; \space \space m\neq n$$

Let's use our old faithful example of particle in a box wavefunctions.

$$\psi_n(x)={(\frac{2}{a})}^{1/2}sin\frac{n \pi x}{a}$$

For this we'll need our trigonometric identities, specifically:

$$sin\alpha \space sin\beta = \frac{1}{2}cos(\alpha - \beta) - \frac{1}{2}cos(\alpha + \beta)$$

$$\frac{2}{a}\int^{a}_{0}sin\frac{n \pi x}{a}sin\frac{m \pi x}{a}dx = \frac{1}{a}cos\frac{(n - m)\pi x}{a}dx - \frac{1}{a}\int^{a}_{0}cos\frac{(n + m)\pi x}{a}dx$$

Since m and n are integers, both integrands go through half cycles of $\pi$, so both integrands are 0 when $m\neq n$. Therefore,

$$\frac{2}{a}=\int^a_0sin\frac{n \pi x}{a}sin\frac{m \pi x}{a}dx = 0; \space \space m\neq n$$

Let's visually look at how integrands containing $sin$ and $cos$ work.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y2, label='cos(x)')
plt.plot(x, y, label='sin(x)')
plt.hlines(0, 0, 2*np.pi, colors='r', linestyles='dashed')

# x axis in units of pi/2
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])

#shade area from cos to 0 until pi
plt.fill_between(x, y2, 0, where=(x < np.pi), color='blue', alpha=0.3)
#shade area from sin to 0 until pi
plt.fill_between(x, y, 0, where=(x < 2 * np.pi), color='orange', alpha=0.3)
```

From this we can see that integrals of $cos x$ vanish when the limits are half cycles of $cos x$ and for integrals with $sin x$ they vanish when the limits are full cycles of $sin x$.

The combined requirements of orthogonality and normalization of our wavefunctions is called orthonormality. We define this succintly as 

$$\int^{\infty}_{\infty}\psi_m^*(x)\psi_n(x)dx = \delta_{mn}$$

where $\delta_{mn}$ is called the Kroenecker delta defined as

$$
\delta_{mn} = \begin{cases}
    1 \space \space i=j \\
    0 \space \space i\neq j
    \end{cases}
$$


Now, let's talk about why quantum mechanical operators must be real.

The property is defined by the equation where

$$\int_{-\infty}^{\infty}f^*(x)\hat{A}g(x)dx = \int_{-\infty}^{\infty}g(x){[Af]}^*(x)dx$$

This depicts $\hat{A}$ operating on $g(x)$ on the left side and $\hat{A}^*$ operating on $f^*(x)$ on the right hand side.

For this example we will use

$$
\begin{align}
    f(x) &= \frac{1}{\pi^{1/4}}e^{-x^2/2} \space \space -\infty < x < \infty \\

    g(x) &= \frac{2^{1/2}}{\pi^{1/4}}e^{-x^2/2} \space \space -\infty < x < \infty \\

    \hat{A}&=\hat{P_x}=-i\hbar\frac{d}{dx}
\end{align}
$$


Let's calculate $\hat{A}g(x)$

$$
\begin{align}
\hat{A}g(x)&=-i\hbar\frac{d}{dx}\frac{2^{1/2}}{\pi^{1/4}}xe^{-x^2/2} \\
& = -i\hbar\frac{2^{1/2}}{\pi^{1/4}}[e^{-x^2/2} - x^2e^{-x^2/2}]
\end{align}
$$

and plug it in:

$$
\begin{align}
\int^{\infty}_{-\infty}f^*(x)\hat{A}g(x)dx & =-i\hbar{(\frac{2}{\pi})}^{1/2}\int^{\infty}_{-\infty}(e^{-x^2} - x^2e^{-x^2/2})dx\\

&= -i\hbar{(\frac{2}{\pi})}^{1/2}(\pi^{1/2} - \frac{\pi^{1/2}}{2})\\
&= -\frac{i\hbar}{2^{1/2}}

\end{align}
$$

Now $\hat{A}^*f(x)$

$$
\begin{align}
\hat{A}^*f(x)&= i\hbar \frac{d}{dx}\frac{1}{\pi^{1/4}}e^{-x^2/2}\\
& = -\frac{i\hbar}{\pi^{1/4}}xe^{-x^2/2}
\end{align}
$$

Now, let's plug that in.

$$
\begin{align}

\int^{\infty}_{-\infty}g(x)\hat{A}^*f^*(x)dx & =-i\hbar{(\frac{2}{\pi})}^{1/2}\int^{\infty}_{-\infty}x^2e^{-x^2/2}dx\\
&= -i\hbar{(\frac{2}{\pi})}^{1/2}\cdot\frac{\pi^{1/2}}{2}\\
&= -\frac{i\hbar}{2^{1/2}}

\end{align}
$$


We get the same result, so $\hat{P_x}$ satisfies our definition. Thus, it is said that $\hat{P_x}$ is said to be *Hermitian*.

## Commutation Relationships

in normal algebraic expressions, quantities commute. This is not always the case in quantum mechanics.

Two commuting quantum mechanical operators are defined as:

$$\hat{A}\hat{B}f(x)=\hat{B}\hat{A}f(x)$$

whereas two noncummative operators are defined as:

$$\hat{A}\hat{B}f(x)\neq\hat{B}\hat{A}$$

A succint way of writing these relations is $[\hat{A}, \hat{B}]$ which corresponds to $\hat{A}\hat{B} - \hat{B}\hat{A}$. If two operators commute, this quantity is 0, otherwise they do not commute.

One manner in which this is useful, is the fact that if two operators commute they can be measured ***simultaneously*** to ***any*** precision. Otherwise, this is where the Uncertainty Principle emerges. Here's a more rigourous expression for the Uncertainty Principle.

$$\sigma_a\sigma_b\geq \frac{1}{2}\biggl\lvert{\int\psi^*(x)[\hat{A},\hat{B}]\psi(x)dx}\biggr\rvert$$

Let's compute this for our $\hat{P_x}$ and $\hat{X}$ operators.

$$
\begin{align}
[\hat{P_x}, \hat{X}] &= -i\hbar\frac{d(x)}{dx} + x i \hbar \frac{d(1)}{dx}\\
&= -i\hbar\cdot 1 + 0\\
&= -i\hbar
\end{align}
$$


Do they commute? No, so can we measure them each simultaneously to an arbitrary precision? Also, **NO**!

Now, that we have that relationship, let's plug that into our definition of the Uncertainty Principle.

$$
\begin{align}
\sigma_p\sigma_x &\geq \frac{1}{2}\biggl\lvert\int\psi^*(x)(-i\hbar)\psi(x)dx\biggr\rvert\\
&\geq \frac{1}{2}\lvert-i\hbar\rvert\geq\frac{\hbar}{2}
\end{align}
$$
